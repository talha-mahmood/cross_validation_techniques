```python
import pandas as pd

df=pd.read_csv('adult.csv')

df.head()
```

```python
df.isnull().sum()
```

```python
df['sex']=df['sex'].map({'Male':0,'Female':1})
```

```python
df['marital.status'].unique()
```

```python
df['marital.status']=df['marital.status'].replace(['Widowed','Divorced','Separated','Never-married'],0)
df['marital.status']=df['marital.status'].replace(['Married-civ-spouse','Married-spouse-absent','Married-AF-spouse'],1)
df.head()
```

```python
df.drop(labels=['workclass','education','occupation','relationship','race','native.country'],axis=1,inplace=True)
```

```python
df.head(10)
```

```python
y=df['income']
x=df.drop(labels=['income'],axis=1)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

models=[]
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
models.append({'LR',LogisticRegression()})
models.append({'KNN',KNeighborsClassifier()})
```

```python
import numpy as np
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score,StratifiedKFold, LeaveOneOut, RepeatedKFold
results=dict()
for name,model in models:
  kFold=KFold(n_splits=10,random_state=7,shuffle=True)
  cv_results=cross_val_score(model,x_train,y_train,cv=kFold,scoring='accuracy')
  results[name]=(cv_results.mean(),cv_results.std())

print()
print("result  result.mean    result.std")
for name,result in results.items():
  print(name,result)
```

```python
results=dict()
for name,model in models:
  stratifiedkFold=StratifiedKFold(n_splits=10,random_state=7,shuffle=True)
  cv_results=cross_val_score(model,x_train,y_train,cv=stratifiedkFold,scoring='accuracy')
  results[name]=(cv_results.mean(),cv_results.std())

print()
print("result  result.mean    result.std")
for name,result in results.items():
  print(name,result)
```

```python
results=dict()
for name,model in models:
  repeatedkFold=RepeatedKFold(n_splits=10, n_repeats=3, random_state=7)
  cv_results=cross_val_score(model,x_train,y_train,cv=repeatedkFold,scoring='accuracy')
  results[name]=(cv_results.mean(),cv_results.std())

print()
print("result  result.mean    result.std")
for name,result in results.items():
  print(name,result)
```
